```python
import matplotlib.pyplot as lib
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

```python
sns.distplot(crash_df["not_distracted"], kde=True, bins = 30)
```

```python
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind = "kde")
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind = "reg")
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind = "hex")
```

```python
sns.kdeplot(crash_df["alcohol"])
```

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df,hue = "sex", palette= "Reds")
```

**Styling**

```python
sns.set_style("darkgrid")
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind = "reg")
```

```python
sns.set_style("white")
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind = "reg")
```

```python
sns.set_style("whitegrid")
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind = "reg")
```

```python
sns.set_style("ticks")
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind = "reg")
```

```python
sns.set_style("ticks")
sns.set_context("paper", font_scale = 1.4)
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind = "reg")
```

```python
sns.set_style("ticks")
sns.set_context("poster", font_scale = 1.4)
sns.jointplot(x = "speeding", y = "alcohol", data = crash_df, kind = "reg")
```
